## Topology A


```python
picklefolder = '../../../data/results/picklejar/'
with gzip.open(picklefolder + 'beetle-topA.pgz', "rb") as f:
    hmc_traces = cloudpickle.load(f, encoding='bytes')

hmc_trace01=hmc_traces['trace01']
hmc_trace02=hmc_traces['trace02']
hmc_trace03=hmc_traces['trace03']
hmc_trace04=hmc_traces['trace04']
hmc_trace05=hmc_traces['trace05']

hmc_trace15=hmc_traces['trace15']
hmc_trace3=hmc_traces['trace3']
hmc_trace5=hmc_traces['trace5']
hmc_trace7=hmc_traces['trace7']
hmc_trace100=hmc_traces['trace100']
```

```python
hmc_traces = [
    hmc_trace01,
    hmc_trace02,
    hmc_trace03,
    hmc_trace04,
    hmc_trace05,
    hmc_trace15,
    hmc_trace3,
    hmc_trace5,
    hmc_trace7,
    hmc_trace100,
]
```

```python
# Suppress RuntimeWarnings only
warnings.filterwarnings("ignore", category=RuntimeWarning)


for i, trace in enumerate(hmc_traces):

    print("trace", i)

    # Summary with ESS and R-hat
    summary = az.summary(trace, round_to=2)
    # print(summary)

    summary = summary[~summary.index.str.contains("vn")]
    print("ess bulk:", len(summary[summary['ess_bulk']<500]))
    print("ess tail:", len(summary[summary['ess_tail']<500]))
    print("r hat:", len(summary[(summary['r_hat']!=1) & (summary['r_hat']!=1.01)]))

    # Check for divergences
    #az.plot_energy(trace)
    #az.plot_trace(trace)

    # Divergence count
    n_divergent = trace.sample_stats["diverging"].sum().values
    print("Number of post-warmup divergences:", n_divergent)
```
